```python
import matplotlib.pyplot as plt
import numpy as np
from cvxopt import matrix, solvers
import CQE_learner
import Foe_learner
import Friend_learner
import Q_learner
import Soccer
```

```python
QL1 = Q_learner.QLearner(env=Soccer.SoccerGame(train=False),lin=False)
QL1.train_loop()
```

```python
QL_error1 = np.array(QL1.error_log)
QL_it_num1 = np.array(QL1.iter_list)
plt.figure(figsize=(10,7))
plt.plot(QL_it_num1,QL_error1,linewidth=1)
plt.plot(QL_it_num1,QL1.alpha_schedule[QL_it_num1])
plt.title('Q Learner')
plt.ylabel('Q-value Differnce')
plt.xlabel('Iteration Number')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0, 0))
plt.ylim((0,0.5))
plt.show()
```
